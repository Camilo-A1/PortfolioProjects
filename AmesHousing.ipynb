## Modelo PCA

```python
# Se normalizan los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_top)
```

```python
# Se ajusta el modelo PCA
modelo_pca = PCA(n_components=6)
modelo_pca.fit(scaled_data)
```

# Interpretación

Una vez finalizado el entrenamiento del objeto PCA, se puede acceder a toda la información de las componentes creadas.

El atributo "components_" almacena los valores de los loadings ϕ que definen cada componente (eigenvector). Las filas representan las componentes principales, ordenadas por su varianza explicada de mayor a menor, y las columnas corresponden a las variables de entrada.

```python
# Se convierte el array a dataframe para añadir nombres a los ejes.
pd.DataFrame(
    data    = modelo_pca.components_,
    columns = df_top.columns
)
```

Es posible analizar visualmente la influencia de las variables en cada componente utilizando un gráfico de tipo heatmap.

```python
# Heatmap componentes
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
componentes = modelo_pca.components_
plt.imshow(componentes.T, cmap='viridis', aspect='auto')
plt.yticks(range(len(df_top.columns)), df_top.columns)
plt.xticks(range(6), np.arange(modelo_pca.n_components_) + 1)
plt.grid(False)
plt.colorbar();
```

Una vez que se han calculado los componentes principales, es posible obtener información sobre la varianza explicada por cada una de ellas, así como la proporción que representan respecto al total y la proporción acumulada de varianza. Esta información se encuentra almacenada en los atributos explained_variance_ y explained_variance_ratio_ del modelo.

```python
# Porcentaje de varianza explicada por cada componente
print('Porcentaje de varianza explicada por cada componente: \n')
print(modelo_pca.explained_variance_ratio_)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.bar(
    x      = np.arange(modelo_pca.n_components_) + 1,
    height = modelo_pca.explained_variance_ratio_
)

for x, y in zip(np.arange(len(df_top.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
```

En este caso, el primer componente explica el 21% de la varianza observada en los datos y el segunda el 8%.

```python
# Porcentaje de varianza explicada acumulada
prop_varianza_acum = modelo_pca.explained_variance_ratio_.cumsum()
print('Porcentaje de varianza explicada acumulada: \n')
print(prop_varianza_acum)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.plot(
    np.arange(6) + 1,
    prop_varianza_acum,
    marker = 'o'
)

for x, y in zip(np.arange(len(df_top.columns)) + 1, prop_varianza_acum):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_ylim(0, 1.1)
ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_title('Porcentaje de varianza explicada acumulada')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza acumulada');
```

Si se empleasen 4 componentes se conseguiría explicar el 95% de la varianza observada.

### Regresión Lineal

```python
# Necesitamos escalar antes de hacer la transformación
scaler = StandardScaler()
df_top.loc[:, df_top.columns != 'SalePrice'] = scaler.fit_transform(df_top.loc[:, df_top.columns != 'SalePrice'])
```

```python
# Dividimos en set de entrenamiento y de prueba. Conjunto de pruebas equivalente al 20 % de los datos
X_train, X_test, y_train, y_test = train_test_split(df_top.drop('SalePrice', axis=1), df_top[['SalePrice']], test_size=0.2, random_state=42)

print('Cantidad de datos para train: ', len(X_train))
print('Cantidad de datos para test: ', len(X_test))
```

```python
# La figura de train
X_train.shape
```

```python
# La figura de shape
X_test.shape
```

```python
# Ajuste del modelo de regresión lineal
modelo = LinearRegression().fit(X_train, y_train)
```

```python
# Predicción de valores en el conjunto de prueba
y_pred = modelo.predict(X_test)
```

```python
# Cálculo del error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", mse)
```

Que el MSE sea 1036872522.9056956 significa que en promedio, las predicciones del modelo están alejadas de los valores reales de la variable objetivo en una cantidad que corresponde a la raíz cuadrada de este número.

### Máquina de vectores de soporte de margen suave (SVM)
Para ejecutar el análisis se utilizarán las librerías `sklearn.svm` y `sklearn.metrics` y se probarán tres tipos de modelos SVR (lineal, radial y polinómico).

```python
# Se importan las librerías necesarias
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
```

```python
# Se entrena y ajusta el modelo utilizando cada opción de kernel
op1 = SVR(kernel='linear')
op1.fit(X_train, y_train)
y_pred_op1 = op1.predict(X_test)

op2 = SVR(kernel='rbf')
op2.fit(X_train, y_train)
y_pred_op2 = op2.predict(X_test)

op3 = SVR(kernel='poly')
op3.fit(X_train, y_train)
y_pred_op3 = op3.predict(X_test)
```

```python
# Se calcula el MAE para cada modelo y se selecciona el modelo con el MAE más bajo
mae_op1 = mean_absolute_error(y_test, y_pred_op1)
mae_op2 = mean_absolute_error(y_test, y_pred_op2)
mae_op3 = mean_absolute_error(y_test, y_pred_op3)

print("MAE para opción 1 (kernel lineal):", mae_op1)
print("MAE para opción 2 (kernel radial):", mae_op2)
print("MAE para opción 3 (kernel polinómico):", mae_op3)

mejor_opcion = min(mae_op1, mae_op2, mae_op3)
print("La mejor opción es:", end="")
if mejor_opcion == mae_op1:
    print(" 1 (kernel lineal)")
elif mejor_opcion == mae_op2:
    print(" 2 (kernel radial)")
else:
    print(" 3 (kernel polinómico)")
```

El modelo SVR con kernel lineal logró predecir la variable objetivo con un error medio absoluto (MAE) menor que los otros dos modelos SVR con kernels radial y polinómico. Esto indica que el modelo SVR con kernel lineal es capaz de ajustarse mejor a los datos de entrenamiento y generalizar de manera más precisa a los datos de prueba.
